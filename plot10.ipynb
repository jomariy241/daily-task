```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

import pandas as pd
df = pd.read_excel('/content/Superstore.xls')
df.head()
```

```python
#scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(df['Sales'], df['Profit'], c='blue', edgecolors='w')
plt.title('Profit on Sales')
plt.xlabel('Sales')
plt.ylabel('Profit')
plt.show()
```

```python
plt.hist(df['Sales'], color='green', rwidth=0.9)
plt.title('Sales Distribution', fontsize=16)
plt.xlabel('Sales', fontsize=14)
plt.ylabel('Frequency', fontsize=14)
plt.show()
```

```python
#bar plot
plt.bar(df['Country'],df['Sales'])
plt.title('sales')
plt.xlabel('Country')
plt.ylabel('Sales')
```

```python
#heatmap
data=df[['Sales','Quantity','Profit']]
sns.heatmap(data.corr(),annot=True)
```

```python
# box plot
plt.figure(figsize=(12, 8))
sns.boxplot(x='Category', y='Sales', data=df)
plt.title('Box Plot of Sales by Category', fontsize=16)
plt.xlabel('Category', fontsize=14)
plt.ylabel('Sales', fontsize=14)
plt.grid(True)
plt.show()
```
